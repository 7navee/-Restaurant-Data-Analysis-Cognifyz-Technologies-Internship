#Task 1 Top Cuisines
Determine the top three most common cuisines in the dataset?
Calculate the percentage of restaurants that serve each of the top cuisines.

```python
top_cuisines = df['Cuisines'].str.split(', ').explode().value_counts().head(3)
#.str.split() converting each entry into a list of cuisines.
#.explode() This transforms lists into separate rows
#.value_counts() counts the occurrence of unique cuisisnes
total_restaurants = len(df)
top_cuisines_percentage = (top_cuisines / total_restaurants) * 100
```

```python
print("Top 3 Most Common Cuisines:")
print(top_cuisines_percentage)
```

#task 2: City Analysis
Identify the city with the highest number of restaurants in the dataset?

Calculate the average rating for restaurants in each city?

Determine the city with the highest average rating?

```python
city_counts = df['City'].value_counts() #counts the occurrence of each unique city
top_city = city_counts.idxmax() # .idxmax returns the city name which as highest count
```

```python
avg_rating_by_city = df.groupby('City')['Aggregate rating'].mean()
highest_avg_rating_city = avg_rating_by_city.idxmax()
```

```python
print(f"City with the most restaurants: {top_city}")
print(f"City with highest average rating: {highest_avg_rating_city}")
```

Task 2:Votes Analysis
Identify the restaurants with the highest and lowest number of votes.
    
Analyze if there is a correlation between the number of votes and the rating of a restaurant.

```python
#Votes Analysis

most_voted = df[df['Votes'] == df['Votes'].max()]
least_voted = df[df['Votes'] == df['Votes'].min()].head(5)
correlation_votes_rating = df[['Votes', 'Aggregate rating']].corr()
```

```python
print("\nRestaurant with Most Votes:")
print(most_voted[['Restaurant Name', 'Votes']])
print("\nRestaurant with Least Votes:")
print(least_voted[['Restaurant Name', 'Votes']])
print("\nCorrelation between Votes and Ratings:")
print(correlation_votes_rating)
```

```python
most_voted_restaurants = df[['Restaurant Name', 'Votes']].sort_values(by='Votes', ascending=False).head(5)

# most voted restaurants
plt.figure(figsize=(5,3))
sns.barplot(y=most_voted_restaurants['Restaurant Name'], x=most_voted_restaurants['Votes'], palette="Reds")
plt.xlabel("Number of Votes")
plt.ylabel("Restaurant Name")
plt.title("Top 5 Most Voted Restaurants")
plt.show()
```

Task 3:Price Range vs. Online Delivery and Table Booking
Analyze if there is a relationship between the price range and the availability of online delivery and table booking.

Determine if higher-priced restaurants are more likely to offer these services

```python
#Price Range vs. Online Delivery & Table Booking

price_vs_services = df.groupby('Price range')[['Has Online delivery', 'Has Table booking']].value_counts().unstack()
print("\nPrice Range vs. Online Delivery & Table Booking:")
print(price_vs_services)
```

```python
# Analyzing relationships
plt.figure(figsize=(5,3))
sns.boxplot(x=df['Price range'], y=df['Aggregate rating'], hue=df['Has Online delivery'], palette="coolwarm")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.title("Price Range vs Aggregate Rating (Online Delivery)")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x=df['Price range'], y=df['Aggregate rating'], hue=df['Has Table booking'], palette="viridis")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.title("Price Range vs Aggregate Rating (Table Booking)")
plt.show()
```
